```python
# Convert the data to a Pandas dataframe
df = pd.json_normalize(data)
```

```python
# Filter the dataframe to show only rows with occupation_field label 'Data/IT'
filtered_df = df[df['occupation_field.label'] == 'Data/IT']
```

```python
# Save the filtered data to a new JSON file
filtered_data = json.loads(filtered_df.to_json(orient='records', force_ascii=False))
with open('2020_f.json', 'w', encoding='utf-8') as f:
    json.dump(filtered_data, f, ensure_ascii=False, indent=4)
```
